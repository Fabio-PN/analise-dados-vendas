```python
# Criando novas colunas
# Criando coluna de Receitas
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
# Listando as 5 primeiras linhas de dados para conferir a nova coluna Receita
df.head()
```

```python
# Retornando a maior receita 
df["Receita"].max()
```

```python
# Retornando a menor receita
df["Receita"].min()
```

```python
# Retornando as 3 melhores lojas baseado na maior Receita
df.nlargest(3, "Receita")
```

```python
# Retornando as 3 piores lojas baseado na menor Receita
df.nsmallest(3, "Receita")
```

```python
# Agrupamento de vendas por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
# Ordenando o conjunto de dados acima apresentados
# Listar as 10 maiores vendas baseadas nas Receitas do maior para o menor
df.sort_values("Receita", ascending=False).head(10)
```

```python
# Trabalhando com datas
# Transformando o tipo de dado em tipo inteiro
from numpy import int64

df["Data"] = df["Data"].astype(int64)
```

```python
# Verificando o tipo de dado em cada coluna
df.dtypes
```

```python
# Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
# Conferindo se o tipo de dado em Data foi modificado
df.dtypes
```

```python
# Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
# Criando uma nova coluna com o ano
df["Ano Venda"] = df["Data"].dt.year
```

```python
# Coletando uma amostra de 5 linhas de dados
df.sample(5)
```

```python
# Extraindo o mês e o dia e criando duas novas colunas relativas aos mesmos
df["Mês Venda"], df["Dia Venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
# Extraindo uma amostra dos dados após inclusão dos novos campos relativos a mês e dia da venda
df.sample(5)
```

```python
# Retornando a data mais antiga
df["Data"].min()
```

```python
# Calculando a diferença de dias
df["Dif. Dias"] = df["Data"] - df["Data"].min()
```

```python
# Coletando uma amostra de dados
df.sample(5)
```

```python
# Criando uma coluna com o trimestre
df["Trimestre Venda"] = df["Data"].dt.quarter
```

```python
# Coletando uma amostra de dados
df.sample(5)
```

```python
# Filtrando apenas vendas de um mês e ano específico
# Criamos uma variável que irá receber todo o valor filtrado para que seja apresentado em tela
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
# Apresentando os valores filtrados
vendas_marco_19
```

```python
# Apresentando uma amostra dos dados contidos nessa variável
vendas_marco_19.sample(20)
```

```python
# Adicionando um título e alterando o nome dos eixos
import matplotlib.pyplot as plt
from pandas import value_counts
df["Cidade"].value_counts().plot.bar(title="Total de Vendas por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
# Alterando a cor do gráfico
df["Cidade"].value_counts().plot.bar(title="Total de Vendas por Cidade", color="red")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
# Alterando o estilo do gráfico
plt.style.use("ggplot")
```

```python
# Criando um gráfico com o novo estilo
df.groupby(df["Mês Venda"])["Qtde"].sum().plot(title="Total de Produtos Vendidos x Mês")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend();
```

```python
# Agrupando o mês venda e a soma da quantidade de produtos vendidos nesse mês
df.groupby(df["Mês Venda"])["Qtde"].sum()
```

```python
# Selecionando apenas o ano de 2019 para analisar os dados
df_2019 = df[df["Ano Venda"] == 2019]
```

```python
# Total produtos vendidos por mês em 2019
df_2019.groupby(df_2019["Mês Venda"])["Qtde"].sum().plot(marker = "o") # Plotando o gráfico com o marvador em forma de círculo
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend();
```

```python
# Criando um gráfico de histograma
plt.hist(df["Qtde"], color="orangered")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
```

```python
# Criando um gráfico de dispersão
plt.scatter(x=df_2019["Dia Venda"], y=df_2019["Receita"])
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
```

```python
# Salvando os gráficos apresentados em arquivos png
# Agrupar os dados contidos na variável df_2019 associando mês de venda e a soma da quantidade de produtos vendidos
# Apresentar o resultado em um gráfico de linha com marcadores em formato de círculo
df_2019.groupby(df_2019["Mês Venda"])["Qtde"].sum().plot(marker="o") 
plt.title("Quantidade de produtos vendidos x Mês") # # Determinar o título do gráfico
plt.xlabel("Mês")                                  # Determinar a etiqueta no eixo x
plt.ylabel("Total Produtos Vendidos");             # Determinar a etiqueta no eixo y
plt.legend()                                       # Exibir a legenda do gráfico
plt.savefig("Gráfico Qtde x Mês.png")              # Salvar o gráfico em uma figura .png
```
